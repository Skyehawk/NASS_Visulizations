```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
import numpy as np
```

```python
path_econ = "./data/NASS_census_all_receipts_by_year_and_econ_class.csv"
path_opp = "./data/NASS_census_all_operations_by_year_and_econ_class.csv"
df_econ = pd.read_csv(path_econ)
df_opp = pd.read_csv(path_opp)
```

```python
# Clean the 'Value' column
df_econ['Value'] = df_econ['Value'].str.strip()
df_econ['Value'] = df_econ['Value'].str.replace(',', '', regex=False)

# Convert to numeric and handle errors
df_econ['Value'] = pd.to_numeric(df_econ['Value'], errors='coerce')

df_econ
```

```python
# Clean the 'Value' column
df_opp['Value'] = df_opp['Value'].str.strip()
df_opp['Value'] = df_opp['Value'].str.replace(',', '', regex=False)

# Convert to numeric and handle errors
df_opp['Value'] = pd.to_numeric(df_opp['Value'], errors='coerce')

df_opp
```

```python
# Define mapping dictionary for original ranges to new ones
range_mapping = {
    'ECONOMIC CLASS: (LESS THAN 1,000 $)': '\$1,000 - \$9,999',
    'ECONOMIC CLASS: (1,000 TO 2,499 $)': '\$1,000 - \$9,999',
    'ECONOMIC CLASS: (2,500 TO 4,999 $)': '\$1,000 - \$9,999',
    'ECONOMIC CLASS: (5,000 TO 9,999 $)': '\$1,000 - \$9,999',
    'ECONOMIC CLASS: (10,000 TO 24,999 $)': '\$10,000 - \$99,999',
    'ECONOMIC CLASS: (25,000 TO 49,999 $)': '\$10,000 - \$99,999',
    'ECONOMIC CLASS: (50,000 TO 99,999 $)': '\$10,000 - \$99,999',
    'ECONOMIC CLASS: (100,000 TO 249,999 $)': '\$100,000 - \$249,999',
    'ECONOMIC CLASS: (250,000 TO 499,999 $)': '\$250,000 - \$499,999',
    'ECONOMIC CLASS: (500,000 TO 999,999 $)': '\$500,000 - \$999,999',
    'ECONOMIC CLASS: (1,000,000 OR MORE $)': '\$1,000,000 or more'
}

```

```python
# Apply mapping to create new category column
df_econ['Domain Category'] = df_econ['Domain Category'].map(range_mapping)

# Convert Domain Category to categorical with the new order
new_order = ['\$1,000 - \$9,999', '\$10,000 - \$99,999', '\$100,000 - \$249,999', 
             '\$250,000 - \$499,999', '\$500,000 - \$999,999', '\$1,000,000 or more']
df_econ['Domain Category'] = pd.Categorical(df_econ['Domain Category'], categories=new_order, ordered=True)

# Group and aggregate by 'Year' and 'New Category', summing the values
grouped_df_econ = df_econ.groupby(['Year', 'Domain Category'])['Value'].sum().unstack()

grouped_df_econ

```

```python
# Apply mapping to create new category column
df_opp['Domain Category'] = df_opp['Domain Category'].map(range_mapping)

# Convert Domain Category to categorical with the new order
new_order_opp = ['\$1,000 - \$9,999', '\$10,000 - \$99,999', '\$100,000 - \$249,999', 
             '\$250,000 - \$499,999', '\$500,000 - \$999,999', '\$1,000,000 or more']
df_opp['Domain Category'] = pd.Categorical(df_opp['Domain Category'], categories=new_order, ordered=True)

# Group and aggregate by 'Year' and 'New Category', summing the values
grouped_df_opp = df_opp.groupby(['Year', 'Domain Category'])['Value'].sum().unstack()

grouped_df_opp

```

```python
# Exclude the $1,000 - $9,999 category
# grouped_df_econ = grouped_df_econ.drop(columns=['\$1,000 - \$9,999'])
# grouped_df_opp = grouped_df_opp.drop(columns=['\$1,000 - \$9,999'])

# Calculate percentages for each category in each year
percentages_econ = grouped_df_econ.div(grouped_df_econ.sum(axis=1), axis=0) * 100
percentages_opp = grouped_df_opp.div(grouped_df_opp.sum(axis=1), axis=0) * 100




```

```python
# Plotting the data
ax = grouped_df_opp.plot(kind='bar', stacked=True, figsize=(12, 8), cmap="summer")

# Annotate percentages on each category in each bar
for i in range(grouped_df_opp.shape[0]):
    for j in range(grouped_df_opp.shape[1]):
        ax.annotate(f'{percentages_opp.iloc[i, j]:.1f}%', 
                    xy=(i, grouped_df_opp.iloc[i, :j].sum() + grouped_df_opp.iloc[i, j] / 2),
                    ha='center', va='center')

# Define custom formatter for Y-axis tick labels
kfmt = lambda x, pos: "{:g}k".format(x / 1000)
ax.yaxis.set_major_formatter(FuncFormatter(kfmt))

plt.title('Farm Operations by Gross Income over Farm Census Years')
plt.xlabel('Year')
plt.ylabel('Number of Operations')
plt.legend(reversed(ax.legend().legendHandles), reversed(new_order), title='Economic Class (Gross $)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Rotate x-axis tick labels
plt.xticks(rotation=0, ha='center')

plt.show()
```
